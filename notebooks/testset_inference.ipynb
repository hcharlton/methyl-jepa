```python
import polars as pl
```

```python
q = (
    pl.scan_parquet('../data/01_processed/train_sets/pacbio_standard_test.parquet')
    .head(1000000)
    )
df = q.collect()
```

```python
df_true = df.filter(pl.col("label") == True)
df_false = df.filter(pl.col("label") == False)

# Calculate desired counts for 70-30 split
total = int(df.height/2)
print(total)
n_true = int(total * 0.7)
print(n_true)
n_false = total - n_true

# Sample from each group
df_true_sampled = df_true.sample(n=n_true, seed=42)
df_false_sampled = df_false.sample(n=n_false, seed=42)

# Concatenate to get the final DataFrame
result = pl.concat([df_true_sampled, df_false_sampled])

result.head()
```
